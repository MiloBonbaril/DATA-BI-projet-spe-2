```python
import pandas as pd
```

```python
df = pd.read_csv("../data/camelyon17_v1.0/metadata.csv")
```

```python
df.info()
```

```python
# Analyse de la colonne "tumor" qui est la cible
print(df["tumor"].value_counts())
```

```python
# check how many unique patients there are
print("Nombre de patients uniques:", df["patient"].nunique())
```

```python
# analyse de la colonne "split", et de la répartition des échantillons entre les splits
print(df["split"].value_counts())
```

```python
def make_path(row):
    p = f"{row.patient:03d}"
    return (
        f"data/camelyon17_v1.0/patches/"
        f"patient_{p}_node_{row.node}/"
        f"patch_patient_{p}_node_{row.node}_x_{row.x_coord}_y_{row.y_coord}.png"
    )

df["img_path"] = df.apply(make_path, axis=1)
df.head()
```

```python
train_df = df[df["split"] == 0]
val_df = df[df["split"] == 1]

print("Train set size:", len(train_df))
print("répartition tumor train set:", train_df["tumor"].value_counts())
print("\nValidation set size:", len(val_df))
print("répartition tumor validation set:", val_df["tumor"].value_counts())
```

```python
print("Train set, centers distribution:", train_df["center"].value_counts())
print("Validation set, centers distribution:", val_df["center"].value_counts())
```

```python
print("Train set, patients distribution:", train_df["patient"].value_counts())
print("Validation set, patients distribution:", val_df["patient"].value_counts())
```
